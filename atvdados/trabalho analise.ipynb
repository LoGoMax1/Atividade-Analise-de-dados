```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
text_file = pd.read_csv('top50.csv', encoding='utf-8')
```

```python
df = text_file
```

```python
#Qual São os Tipos de gêneros musicais nas músicas fornecidas?
genero_distribuicao = df['Genre'].value_counts()
plt.figure(figsize=(10, 6))
genero_distribuicao.plot(kind='bar')
plt.title('Tipos dos Gêneros Musicais')
plt.xlabel('Gênero')
plt.ylabel('Contagem')
plt.xticks(rotation=45)
plt.show()
```

```python
# 2. Relação entre dançabilidade e energia das músicas (gráfico de dispersão)
plt.figure(figsize=(8, 6))
plt.scatter(df['Danceability'], df['Energy'], color='green', alpha=0.5)
plt.title('Relação entre Dançabilidade e Energia das Músicas')
plt.xlabel('Dançabilidade')
plt.ylabel('Energia')
plt.show()
```

```python
# 3. Distribuição da popularidade das músicas (gráfico de colunas)
popularity_mean = df.groupby('Artist.Name')['Popularity'].mean().sort_values(ascending=False).head(10)
plt.figure(figsize=(10, 6))
popularity_mean.plot(kind='bar')
plt.title('Top 10 Artistas com Maior Popularidade Média')
plt.xlabel('Artista')
plt.ylabel('Popularidade Média')
plt.xticks(rotation=45)
plt.show()
```

```python
# 4. Relação entre dançabilidade e energia das músicas (gráfico de dispersão)

plt.figure(figsize=(8, 6))
plt.scatter(df['Danceability'], df['Energy'], color='green', alpha=0.5)
plt.title('Relação entre Dançabilidade e Energia das Músicas')
plt.xlabel('Dançabilidade')
plt.ylabel('Energia')
plt.show()
```

```python
# 5. Distribuição do volume (loudness) das músicas (indicador)
plt.figure(figsize=(8, 6))
plt.hist(df['Popularity'], bins=20, color='orange')
plt.title('Distribuição da Popularidade das Músicas')
plt.xlabel('Popularidade')
plt.ylabel('Contagem')
plt.show()
```

```python
# 6. Relação entre duração das músicas e popularidade (gráfico de dispersão)
plt.figure(figsize=(8, 6))
plt.scatter(df['Length.'], df['Popularity'], color='purple', alpha=0.5)
plt.title('Relação entre Duração das Músicas e Popularidade')
plt.xlabel('Duração (segundos)')
plt.ylabel('Popularidade')
plt.show()
```

```python
# 7. Distribuição do volume (loudness) das músicas (histograma)
plt.figure(figsize=(8, 6))
plt.hist(df['Loudness..dB..'], bins=20, color='pink')
plt.title('Distribuição do Volume (Loudness) das Músicas')
plt.xlabel('Volume (Loudness)')
plt.ylabel('Contagem')
plt.show()
```
